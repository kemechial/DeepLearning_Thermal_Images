```python
image_size = (480,640)
d_path = "/gdrive/My Drive/Yedek/Thermal_Image/Static_v3/Frontal/train/healthy"
files=next(os.walk(d_path))[2]
```

```python
X_train=np.empty((0,image_size[0],image_size[1],3),dtype='float32')
for file in files:
  file=join(d_path,file)
  img= Image.open(file)
  np_img = np.array(img)
  np_img=np_img.reshape((1,480,640,3))
  X_train=np.concatenate((X_train,np_img),axis=0)
print(X_train.shape)
```
