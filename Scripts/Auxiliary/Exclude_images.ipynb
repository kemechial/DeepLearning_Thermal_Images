```python
def flip_image(image_path,save_path):
  original_img = Image.open(image_path)
  vertical_img = original_img.transpose(method=Image.FLIP_LEFT_RIGHT)
  vertical_img.save(save_path)
  original_img.close()
  vertical_img.close()
```

# **Create a recursive function to walk subdirectories**

```python
def walk_subdirectories(d_path,s_path,view_flag=False):
   subdirectories,files=next(os.walk(d_path))[1:]
   print(d_path,s_path)
   print(subdirectories)
   print(files)
   if(len(files)==0):
      print("No files yet, keep walking...")
      for subdirectory in subdirectories:
        os.mkdir(s_path+'/'+subdirectory)
        if subdirectory not in ['healthy','sick']:
          walk_subdirectories(d_path+'/'+subdirectory,s_path+'/'+subdirectory, subdirectory)
        else:
          walk_subdirectories(d_path+'/'+subdirectory,s_path+'/'+subdirectory, view_flag)
   else:
    print("Files found, checking ids...")
    for file in files:
      file_id=re.split('[A-Z]0*',file.split(".")[0])[1]
      file_id+='_'+file.split(".")[1]
      print(file_id)
      if file_id not in exclusion_dictionary[view_flag]:
        if file_id in revert_dictionary[view_flag]:
          flip_image(d_path+'/'+file, s_path+'/'+f"image_{file_id}.jpg")
          print(f"{file_id} from {view_flag}: image reverted!!!")
        else:
          shutil.copy(d_path+'/'+file, s_path+'/'+f"image_{file_id}.jpg")
      else:
        print(f"{file_id} from {view_flag}: image excluded!!!")
```
