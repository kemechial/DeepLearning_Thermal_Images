```python
directory_path='/gdrive/My Drive/Yedek/Thermal_Image/Static_excluded'
save_path='/gdrive/My Drive/Yedek/Thermal_Image/Static_flipped_v2'
```

```python
def flip_image(image_path,save_path):
  original_img = Image.open(image_path)
  vertical_img = original_img.transpose(method=Image.FLIP_LEFT_RIGHT)
  vertical_img.save(save_path)
  print('image flipped!')
  original_img.close()
  vertical_img.close()
```

```python
def walk_subdirectories(d_path,s_path,flip_flag=False):
   print(flip_flag)
   root,subdirectories,files=next(os.walk(d_path))
   print(f'root:{root},subdirectories:{subdirectories}, number of files: {len(files)}')
   if(len(files)==0):
      print("No files yet, keep walking...")
      for subdirectory in subdirectories:
        os.mkdir(s_path+'/'+subdirectory)
        if subdirectory in ['Left','Left45']:
           walk_subdirectories(d_path+'/'+subdirectory,s_path+'/'+subdirectory,True)
        elif flip_flag:
           walk_subdirectories(d_path+'/'+subdirectory,s_path+'/'+subdirectory,True)
        else:
          walk_subdirectories(d_path+'/'+subdirectory,s_path+'/'+subdirectory)
   else:
     for file in files:
      if flip_flag:
        flip_image(d_path+'/'+file, s_path+'/'+file)
      else:
        shutil.copy(d_path+'/'+file, s_path+'/'+file)
```
