```python
directory_path='/gdrive/My Drive/Yedek/Thermal_Video'
save_path='/gdrive/My Drive/Yedek/Video_Dataset'
```

```python
split=0.3
```

```python
   os.mkdir(save_path+'/train')
   os.mkdir(save_path+'/test')
```

```python
subdirectories=next(os.walk(directory_path))[1]
```

```python
for subdirectory in subdirectories:
   os.mkdir(save_path+'/train/'+subdirectory)
   os.mkdir(save_path+'/test/'+subdirectory)
   files=next(os.walk(directory_path+'/'+subdirectory))[2]
   size=len(files)
   test_size=math.ceil(size*split)
   train_size=size-test_size
   print(f"file number:{size} train:{train_size} test:{test_size}")
   for file in files[:train_size]:
    shutil.copy(directory_path+'/'+subdirectory+'/'+file, save_path+f'/train/{subdirectory}/'+file)
   for file in files[train_size:]:
    shutil.copy(directory_path+'/'+subdirectory+'/'+file, save_path+f'/test/{subdirectory}/'+file)
```
