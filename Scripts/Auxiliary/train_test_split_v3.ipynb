```python
import os
import shutil
import math
```

```python
directory_path='/gdrive/My Drive/Yedek/Thermal_Image/Static_flipped_filtered_v2'
save_path='/gdrive/My Drive/Yedek/Thermal_Image/Static_v3'
```

```python
split=0.3
```

```python
subdirectories=next(os.walk(directory_path))[1]
```

```python
print(subdirectories)
```

```python
for subdirectory in subdirectories:
   os.mkdir(save_path+'/'+subdirectory)
   os.mkdir(save_path+'/'+subdirectory+'/train')
   os.mkdir(save_path+'/'+subdirectory+'/test')
   subdirectories2=next(os.walk(directory_path+'/'+subdirectory))[1]
   for subdirectory2 in subdirectories2:
    os.mkdir(save_path+'/'+subdirectory+f'/train/{subdirectory2}')
    os.mkdir(save_path+'/'+subdirectory+f'/test/{subdirectory2}')
    d_path=directory_path+'/'+subdirectory+'/'+subdirectory2
    files=next(os.walk(d_path))[2]
    size=len(files)
    test_size=math.ceil(size*split)
    train_size=size-test_size
    print(f"file number:{size} train:{train_size} test:{test_size}")
    for file in files[:train_size]:
      shutil.copy(d_path+'/'+ file, save_path+'/'+subdirectory+f'/train/{subdirectory2}/'+file)
    for file in files[train_size:]:
      shutil.copy(d_path+'/'+ file, save_path+'/'+subdirectory+f'/test/{subdirectory2}/'+file)
```
