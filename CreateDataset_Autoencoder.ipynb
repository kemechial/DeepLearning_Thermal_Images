```python
!pip install hickle
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
from google.colab  import drive
drive.mount("/gdrive")
%cd /gdrive
```

```python
img_path = "/gdrive/My Drive/BitirmeTezi/Animals/"
labels=['bird','cat','dog']
```

```python
img_height, img_width = (369,469)
image_size = (200, 200)
batch_size = 32

train_ds, val_ds = tf.keras.utils.image_dataset_from_directory(
    img_path,
    labels='inferred',
    label_mode='int',
    color_mode='rgb',
    batch_size=batch_size,
    image_size=image_size,
    shuffle=True,
    seed=1,
    validation_split=0.3,
    subset='both',
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False,
)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
X_train=np.empty((0,image_size[0],image_size[1],3),dtype='float32')
y_train=np.empty((0,),dtype='float32')
X_test=np.empty((0,image_size[0],image_size[1],3),dtype='float32')
y_test=np.empty((0,),dtype='float32')

print(X_train.shape)
print(y_train.shape)
X_train
```

```python
y_train
```

```python
for image_batch, labels_batch in train_ds:
 X_train=np.concatenate((X_train,image_batch),axis=0)
 y_train=np.concatenate((y_train,labels_batch),axis=0)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
for image_batch, labels_batch in val_ds:
 X_test=np.concatenate((X_test,image_batch),axis=0)
 y_test=np.concatenate((y_test,labels_batch),axis=0)
```

```python
X_test.shape
```

```python
import hickle as hkl
data = {'xtrain': X_train, 'xtest': X_test,'ytrain': y_train,'ytest':y_test}
hkl.dump(data,'/gdrive/My Drive/BitirmeTezi/image_dataset.hkl')
```
