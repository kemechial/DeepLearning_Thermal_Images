```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,
                                    fmax=8000)
```

```python
S.dtype
```

```python
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax,cmap='jet')
images_dir = '/gdrive/My Drive/BitirmeTezi/'
plt.axis('off')
plt.savefig(f"{images_dir}/out.jpg",bbox_inches='tight',pad_inches=0)
plt.axis('on')
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Mel-frequency spectrogram')
```
