```python
img_path = "/gdrive/My Drive/BitirmeTezi/Animals/"
```

```python
labels=['bird','cat','dog']
```

```python
import pandas as pd
import os
audio_list = []
label_list = []
for label in labels:
    for audio_file in os.listdir(img_path+label):
        audio_list.append(img_path+label+'/'+audio_file)
        label_list.append(label)

df = pd.DataFrame({'img':audio_list, 'label':label_list})
```
